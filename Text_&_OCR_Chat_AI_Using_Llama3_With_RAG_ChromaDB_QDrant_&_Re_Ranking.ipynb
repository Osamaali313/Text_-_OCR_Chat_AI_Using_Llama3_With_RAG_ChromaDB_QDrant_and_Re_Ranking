```python
%%time

# Load the Generative model and tokenizer
gen_model_id = '/root/.cache/kagglehub/models/metaresearch/llama-3/transformers/8b-chat-hf/1'

def load_model_and_tokenizer(gen_model_id):
    """
    Load the model and tokenizer with the specified configuration.

    Args:
        gen_model_id (str): Path to the model

    Returns:
        model (transformers.AutoModelForCausalLM): Loaded model
        tokenizer (transformers.AutoTokenizer): Tokenizer for the model
    """
    bnb_config = transformers.BitsAndBytesConfig(
        load_in_8bit=True,  # Load the model in 8-bit quantization
    )

    model_config = transformers.AutoConfig.from_pretrained(gen_model_id)
    model = transformers.AutoModelForCausalLM.from_pretrained(
        gen_model_id,
        trust_remote_code=True,
        config=model_config,
        quantization_config=bnb_config,
        device_map='auto',  # Automatically distribute the model across available devices
    )
    tokenizer = transformers.AutoTokenizer.from_pretrained(gen_model_id)

    return model, tokenizer

try:
    llama3_model, llama3_tokenizer = load_model_and_tokenizer(gen_model_id)
except ImportError as e:
    print("An import error occurred. Please make sure you are running this code with a GPU.")
    print("Error details:", e)
```

```python
# Define a list of markers to check for in the response
markers = ["!assistant", "AI Assistant.", ")assistant", ".assistant", ". More", ". Read more", ".read more", ". more.", ".more.", "...", " Last updated:", " [2] ", ") ("]

def truncate_response(prompt, response, markers):
    # Initialize variables to store the earliest marker and its index
    earliest_marker = None
    earliest_index = len(response)

    # Iterate over the markers and find the earliest occurrence
    for marker in markers:
        index = response.find(marker)
        if index != -1 and index < earliest_index:
            earliest_marker = marker
            earliest_index = index

    # Check for the repeated question
    question_repeat_index = response.find(prompt, response.find(prompt) + len(prompt))

    if question_repeat_index != -1:
        # If the question is repeated twice, truncate the response up to the first occurrence
        response = response[:question_repeat_index].strip()
    elif earliest_marker:
        # If any marker is found, truncate the response up to that point and 1
        response = response[:earliest_index + 1].strip()

    return response
```

```python
class Llama3(CustomLLM):
    num_output: int = 512  # Default number of output tokens
    model_name: str = "Llama3"  # Name of the custom LLM
    model: Any = None  # Placeholder for the loaded model
    tokenizer: Any = None  # Placeholder for the tokenizer

    def __init__(self, model, tokenizer, num_output):
        super(Llama3, self).__init__()
        self.model = model  # Initialize the model
        self.tokenizer = tokenizer  # Initialize the tokenizer
        self.num_output = num_output  # Initialize the number of output tokens

    @property
    def metadata(self) -> LLMMetadata:
        """Get LLM metadata."""
        return LLMMetadata(
            num_output=self.num_output,
            model_name=self.model_name,
        )

    @llm_completion_callback()  # Decorator for completion method
    def complete(self, prompt: str, **kwargs: Any) -> CompletionResponse:
        """
        Generate a completion response for the given prompt.

        Args:
            prompt (str): The input prompt
            **kwargs: Additional keyword arguments

        Returns:
            CompletionResponse: The generated completion response
        """
        formatted_prompt = self._get_prompt(prompt, **kwargs)  # Format the prompt
        input_ids = self.tokenizer.encode(formatted_prompt, return_tensors="pt").to(self.model.device)  # Tokenize the prompt
        attention_mask = torch.ones(input_ids.shape, dtype=torch.long, device=self.model.device)  # Create attention mask
        pad_token_id = self.tokenizer.eos_token_id  # Get the end-of-sequence token ID

        output = self.model.generate(
            input_ids,
            attention_mask=attention_mask,
            pad_token_id=pad_token_id,
            max_length=2048,  # Set the maximum length of generated text
            num_return_sequences=1,  # Generate only one sequence
            do_sample=True,  # Enable sampling from the probability distribution
            temperature=0.4,  # Adjust the temperature for less creative responses
        )  # Generate text based on the prompt

        generated_text = self.tokenizer.decode(output[0], skip_special_tokens=True)  # Decode generated text
        return CompletionResponse(text=generated_text)  # Return the completion response

    @llm_completion_callback()  # Decorator for streaming completion method
    def stream_complete(self, prompt: str, **kwargs: Any) -> CompletionResponseGen:
        """
        Generate a streaming completion response for the given prompt.

        Args:
            prompt (str): The input prompt
            **kwargs: Additional keyword arguments

        Yields:
            CompletionResponse: The generated completion response
        """
        formatted_prompt = self._get_prompt(prompt, **kwargs)  # Format the prompt
        input_ids = self.tokenizer.encode(formatted_prompt, return_tensors="pt").to(self.model.device)  # Tokenize the prompt
        attention_mask = torch.ones(input_ids.shape, dtype=torch.long, device=self.model.device)  # Create attention mask
        pad_token_id = self.tokenizer.eos_token_id  # Get the end-of-sequence token ID

        for output in self.model.generate(
            input_ids,
            attention_mask=attention_mask,
            pad_token_id=pad_token_id,
            max_length=2048,  # Set the maximum length of generated text
            num_return_sequences=1,  # Generate only one sequence
            do_sample=True,  # Enable sampling from the probability distribution
            temperature=0.4,  # Adjust the temperature for less creative responses
        ):  # Generate text based on the prompt
            generated_text = self.tokenizer.decode(output, skip_special_tokens=True)  # Decode generated text
            yield CompletionResponse(text=generated_text, delta=generated_text)  # Yield the completion response

Settings.llm = Llama3(llama3_model, llama3_tokenizer, 512)  # Set the custom LLM with the loaded model, tokenizer, and num_output
```

```python
# Getting the Embedding Model (Both options are open-sourced from Huggingface)
embedding_model_name ="BAAI/bge-large-en-v1.5"
# embedding_model_name = "sentence-transformers/all-mpnet-base-v2" # This works too
encoder = SentenceTransformer(embedding_model_name, device=device)
embeddings = HuggingFaceEmbeddings(model_name=embedding_model_name, model_kwargs=model_kwargs, encode_kwargs={'normalize_embeddings': True})
embedding_tokenizer = AutoTokenizer.from_pretrained(embedding_model_name)
```

```python
model = T5ForConditionalGeneration.from_pretrained("Voicelab/vlt5-base-keywords")
tokenizer = AutoTokenizer.from_pretrained("Voicelab/vlt5-base-keywords")

def extract_keyphrases(text):
    # Extract keyphrases using the T5 extractor
    task_prefix = "Keywords: "
    input_sequences = [task_prefix + text]
    input_ids = tokenizer(
        input_sequences, return_tensors="pt", truncation=True, max_length=512
    ).input_ids

    # Generate keyphrases using the T5 model
    output = model.generate(
        input_ids,
        no_repeat_ngram_size=3,  # avoid repeating any trigrams (sequences of 3 tokens) that already exist in the generated output
        num_beams=5,  # model will use beam search with 5 beams (higher value of num_beams allows the model to consider more possible sequences and can lead to better quality output, but it also increases the computational cost and generation time)
        max_length=50  # Specify the maximum length of the generated output
    )

    predicted = tokenizer.decode(output[0], skip_special_tokens=True)
    keyphrases = predicted.split(', ')
    return keyphrases

def process_keyphrases(keyphrases):
    processed_keyphrases = []
    for keyphrase in keyphrases:
        words = keyphrase.split()
        # Remove stop words if they are the first or last word
        if words and words[0].lower() in stop_words:
            words = words[1:]
        if words and words[-1].lower() in stop_words:
            words = words[:-1]
        # Convert remaining words to uppercase
        processed_keyphrase = ' '.join(word.upper() for word in words)
        if processed_keyphrase:
            processed_keyphrases.append(processed_keyphrase)
    return processed_keyphrases
```

```python
def split_text_into_chunks_with_overlapping_context(text, max_tokens=300, overlap_sentences=2):
    sentences = sent_tokenize(text)
    chunks = []
    current_chunk = []
    for i in range(len(sentences)):
        current_chunk.append(sentences[i])
        if len(embedding_tokenizer.encode(' '.join(current_chunk))) >= max_tokens or i == len(sentences) - 1:
            overlap = ' '.join(current_chunk[-overlap_sentences:])
            chunks.append(' '.join(current_chunk))
            if ' '.join(current_chunk) == overlap:
                overlap = ' '.join(current_chunk[-(overlap_sentences-1):])
                current_chunk = [overlap]
            else:
                current_chunk = [overlap]
    return chunks

def split_and_add_metadata(documents, year):
    split_docs = []
    all_keyphrases = []

    for doc in documents:
        # Split the document into chunks with overlapping context
        splits = split_text_into_chunks_with_overlapping_context(doc.page_content)
        for split in splits:
            # Extract and process keyphrases for each split
            keyphrases = extract_keyphrases(split)
            processed_keyphrases = process_keyphrases(keyphrases)

            # Add the processed keyphrases to the list
            all_keyphrases.extend(processed_keyphrases)

            # Append a Document object with the split text and metadata (year and keyphrases)
            split_docs.append(Document(
                page_content=split,
                metadata={"year": year, "keyphrases": processed_keyphrases}
            ))

    return split_docs, all_keyphrases

def load_and_split_text(file_path, year):
    loader = TextLoader(file_path, encoding="utf8")
    documents = loader.load()

    # Clean up the text by replacing line breaks with spaces if any
    for doc in documents:
        doc.page_content = doc.page_content.replace("\n\n", " ")

    return split_and_add_metadata(documents, year)

# Load and split text for each dataset
datasets = [
    ("/content/president-bidens-sotu-address-2022-2024/presidentbiden-sotu-2022.txt", 2022),
    ("/content/president-bidens-sotu-address-2022-2024/presidentbiden-sotu-2023.txt", 2023),
    ("/content/president-bidens-sotu-address-2022-2024/presidentbiden-sotu-2024.txt", 2024)
]

all_documents = []
all_keyphrases = []

for file_path, year in datasets:
    split_docs, keyphrases = load_and_split_text(file_path, year)
    all_documents.extend(split_docs)
    all_keyphrases.extend(keyphrases)

def print_document_counts(datasets):
    count_before = defaultdict(int)
    count_after = defaultdict(int)

    for file_path, year in datasets:
        loader = TextLoader(file_path, encoding="utf8")
        documents = loader.load()
        count_before[year] = len(documents)

        split_docs, _ = load_and_split_text(file_path, year)
        count_after[year] = len(split_docs)

    total_count_before = sum(count_before.values())
    total_count_after = sum(count_after.values())

    # Print counts
    print("Count of documents before split:")
    for year, count in count_before.items():
        print(f"{year}: {count}")
    print("Total:", total_count_before)

    print("\nCount of chunks after split:")
    for year, count in count_after.items():
        print(f"{year}: {count}")
    print("Total:", total_count_after)

# Print the count of documents before and after splitting
print_document_counts(datasets)
```

```python
def plot_top_keyphrases(all_keyphrases, top_n=25):
    # Count the frequency of each keyphrase
    keyphrase_counts = Counter(all_keyphrases)

    # Get the top N most frequent keyphrases and their counts
    top_keyphrases = keyphrase_counts.most_common(top_n)

    # Separate the keyphrases and counts into separate lists
    keyphrases, counts = zip(*top_keyphrases)

    # Create a DataFrame for plotting
    df = pd.DataFrame({'Keyphrase': keyphrases, 'Count': counts})

    # Set the style and figure size
    sns.set(style='whitegrid', rc={'figure.figsize': (12, 8)})

    # Create a vertical bar chart using Seaborn
    ax = sns.barplot(x='Keyphrase', y='Count', data=df, palette='viridis')

    # Set the chart title and labels
    ax.set_title(f'Top {top_n} Most Frequent Topics', fontsize=16)
    ax.set_xlabel('Topic', fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)

    # Rotate the x-tick labels for better readability
    plt.xticks(rotation=45, ha='right')

    # Show the chart
    plt.tight_layout()
    plt.show()

# Plot the top 25 most frequent keyphrases
plot_top_keyphrases(all_keyphrases, top_n=25)
```

```python
def process_metadata(metadata):
    processed_metadata = {}
    for key, value in metadata.items():
        if isinstance(value, list):
            processed_metadata[key] = ', '.join(value)
        elif isinstance(value, (str, int, float, bool)):
            processed_metadata[key] = value
    return processed_metadata

# Process metadata
processed_documents = [
    Document(
        page_content=doc.page_content,
        metadata=process_metadata(doc.metadata)
    )
    for doc in all_documents
]
```

```python
def retrieve_doc_chromadb(query, vector_db, doc_number):
    """
    Retrieve relevant documents from the vector database based on the query and year.

    Args:
        query (str): The user's query.
        vector_db: The vector database object.
        doc_number (int): The number of documents to retrieve.

    Returns:
        list: A list of retrieved document texts.
    """
    # Search for the year in the query using regular expressions
    match = re.search(r'\b(\d{4})\b', query)
    year = match.group(1) if match else None

    # Retrieve relevant results from the vector database
    results = vector_db.similarity_search(query, k=doc_number*10) # To get more results before it gets filtered down

    # Narrow down the results if a year is found in the query (case by case)
    filtered_results = [result for result in results if str(result.metadata.get('year', '')) == year] if year else results

    # Extract the retrieved document texts
    retrieved_documents = [result.page_content for result in filtered_results[:doc_number]]

    return retrieved_documents
```

```python
def retrieve_reranked_doc_chromadb(query, vector_db, doc_number, reranker):
    """
    Retrieve relevant documents from the vector database based on the query and year.

    Args:
        query (str): The user's query.
        vector_db: The vector database object.
        doc_number (int): The number of documents to retrieve.

    Returns:
        list: A list of retrieved document texts.
    """
    # Search for the year in the query using regular expressions
    match = re.search(r'\b(\d{4})\b', query)
    year = match.group(1) if match else None

    # Retrieve relevant results from the vector database
    results = vector_db.similarity_search(query, k=doc_number*10) # To get more results before it gets filtered down

    # Narrow down the results if a year is found in the query (case by case)
    filtered_results = [result for result in results if str(result.metadata.get('year', '')) == year] if year else results

    # Compute relevance scores for the retrieved documents
    pairs = [[query, result.page_content] for result in filtered_results]
    scores = reranker.compute_score(pairs, normalize=True)

    # Sort the documents based on the relevance scores - higher the score, the greater the relevance
    # https://github.com/FlagOpen/FlagEmbedding/blob/master/FlagEmbedding/reranker/README.md
    sorted_results = sorted(zip(filtered_results, scores), key=lambda x: x[1], reverse=True)

    # Extract the retrieved document texts and their relevance scores
    retrieved_documents = [(result.page_content, score) for result, score in sorted_results[:doc_number]]

    return retrieved_documents
```

```python
def retrieve_doc_qdrant(query, qdrant, collection_name, doc_number):
    """
    Retrieve relevant documents from the Qdrant collection based on the query and year.

    Args:
        query (str): The user's query.
        qdrant: The Qdrant client object.
        collection_name (str): The name of the Qdrant collection.
        doc_number (int): The number of documents to retrieve.

    Returns:
        list: A list of retrieved document texts.
    """
    # Search for the year in the query using regular expressions
    match = re.search(r'\b(\d{4})\b', query)
    year = int(match.group(1)) if match else None

    # Retrieve relevant results from the Qdrant collection
    results = qdrant.search(
        collection_name=collection_name,
        query_vector=encoder.encode(query).tolist(),
        limit=doc_number,
        # Filter by metadata (case by case)
        query_filter=models.Filter(
            must=[
                models.FieldCondition(
                    key="metadata.year",
                    match=models.MatchValue(value=year)
                )
            ]
        ) if year else None
    )

    # Extract the retrieved document texts
    retrieved_documents = [result.payload.get('page_content', '') for result in results[:doc_number]]

    return retrieved_documents
```

```python
def retrieve_reranked_doc_qdrant(query, qdrant, collection_name, doc_number, reranker):
    """
    Retrieve relevant documents from the Qdrant collection based on the query and year.

    Args:
        query (str): The user's query.
        qdrant: The Qdrant client object.
        collection_name (str): The name of the Qdrant collection.
        doc_number (int): The number of documents to retrieve.
        reranker: The reranker object.

    Returns:
        list: A list of tuples containing the retrieved document texts and their relevance scores.
    """
    # Search for the year in the query using regular expressions
    match = re.search(r'\b(\d{4})\b', query)
    year = int(match.group(1)) if match else None

    # Retrieve relevant results from the Qdrant collection
    results = qdrant.search(
        collection_name=collection_name,
        query_vector=encoder.encode(query).tolist(),
        limit=doc_number,
        # Filter by metadata (case by case)
        query_filter=models.Filter(
            must=[
                models.FieldCondition(
                    key="metadata.year",
                    match=models.MatchValue(value=year)
                )
            ]
        ) if year else None
    )

    # Extract the retrieved document texts
    retrieved_documents = [result.payload.get('page_content', '') for result in results]

    # Compute relevance scores for the retrieved documents
    pairs = [[query, doc] for doc in retrieved_documents]
    scores = reranker.compute_score(pairs, normalize=True)

    # Sort the documents based on the relevance scores
    sorted_results = sorted(zip(retrieved_documents, scores), key=lambda x: x[1], reverse=True)

    # Extract the retrieved document texts and their relevance scores
    reranked_documents = [(doc, score) for doc, score in sorted_results[:doc_number]]

    return reranked_documents
```
